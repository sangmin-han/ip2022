```python
import cv2
import numpy as np
```

```python
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('dsu2.jpg')
rows,cols,ch = img.shape
pts1 = np.float32([[181,107],[824,68],[181,604],[824,633]])
pts2 = np.float32([[0,0],[300,0],[0,300],[300,300]])
M = cv2.getPerspectiveTransform(pts1,pts2)
dst = cv2.warpPerspective(img,M,(300,300))
plt.subplot(121),plt.imshow(img),plt.title('Input')
plt.subplot(122),plt.imshow(dst),plt.title('Output1')
plt.show()
```

```python
img = cv2.imread('dsu1.jpg')
rows,cols,ch = img.shape
pts1 = np.float32([[769,218],[1006,251],[780,584],[1006,556]])
pts2 = np.float32([[0,0],[300,0],[0,300],[300,300]])
M = cv2.getPerspectiveTransform(pts1,pts2)
dst = cv2.warpPerspective(img,M,(300,300))
plt.subplot(121),plt.imshow(img),plt.title('Input')
plt.subplot(122),plt.imshow(dst),plt.title('Output2')
plt.show()
```
